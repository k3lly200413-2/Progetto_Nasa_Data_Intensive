The complete catalog of every known Near-Earth Asteroid (NEA) from NASA's Jet Propulsion Laboratory, plus 20 years of close approach data (2015-2035).

### Definizione degli Asteroidi Potenzialmente Pericolosi

La NASA definisce gli Asteroidi Potenzialmente Pericolosi (PHA) sulla base di parametri che misurano il potenziale dell’asteroide di effettuare pericolosi avvicinamenti alla Terra. In particolare, tutti gli asteroidi con una distanza minima di intersezione orbitale (MOID) pari o inferiore a 0,05 au e una magnitudine assoluta (H) pari o inferiore a 22,0, quest’ultima corrispondente approssimativamente a una dimensione superiore ai 140 metri, sono considerati PHA.

spiegazione delle variabili non esplicative per close_approaches:

##### designation:
La designazione può essere temporanea (anno della scoperta, una lettera che indica la quindicina del mese della scoperta dell’oggetto in quell’anno e che va da A (prima metà di gennaio) a Y (seconda metà di dicembre) e un’altra lettera con un suffisso numerico che indica l’ordine di scoperta all’interno di quella quindicina. Le prime 25 scoperte della quindicina ricevono una lettera da A a Z (escludendo la lettera I, che non viene utilizzata per evitare confusione con la cifra 1). Poiché le tecniche moderne producono tipicamente centinaia se non migliaia di scoperte per quindicina, alle scoperte successive all’interno della stessa quindicina viene aggiunto un numero in pedice per indicare quante volte le lettere da A a Z sono già state utilizzate in ciclo. Pertanto il suffisso 1 indica le scoperte dopo un ciclo completo, ossia dalla 26ª alla 50ª scoperta.

La designazione permanente è una sequenza numerica univoca.

<pre> Esempio: designazione provvisoria -> 2005 YQ96 designazione permanente -> 613286 </pre>

##### full_name:
La combinazione di designazioni temporanee e permanenti
<pre>
Esempio:
  613286 (2005 YQ96)
</pre>

##### close_approach_date:
la data e ora del closest approach in date format
<pre>
Example: 2015-01-02 21:46
</pre>

##### distance_au, dist_km e dist_lunar:
La distanza minima in Unità Astronomiche (AU) e Chilometri (km); l’AU è un’unità di lunghezza definita come esattamente pari a 149597870700 m oppure approssimativamente 150 milioni di km; una distanza lunare è la distanza tra la Terra e la Luna: 384399 km.

<pre> Esempio:
    distance_au -> 0.026522 
    dist_km -> 3967664.0 
    dist_lunar -> 10.32 </pre>

##### distance_min_au e distance_max_au
La distanza minima e massima possibile è basata sull’ellisse di errore a 3-sigma del piano bersaglio terrestre. <br>
La distanza minima è la distanza più ravvicinata che l’asteroide potrebbe raggiungere, considerando gli errori attuali nei nostri dati. <br>
La distanza massima è la distanza più lontana a cui l’asteroide potrebbe trovarsi durante quello specifico avvicinamento. <br>

<pre> Esempio:
    l’asteroide 613286 ha distance_au = distance_min_au = distance_max_au. 
    Questo perché siamo stati in grado di calcolare la sua orbita con un errore trascurabile. </pre>



##### velocity_infinity_km_s	

Per gli asteroidi si riferisce alla velocità di un oggetto rispetto a un pianeta (la Terra in questo caso), oppure al Sole, quando è sufficientemente lontano da non essere influenzato dalla sua gravità. <br>
Normalmente $v_{\infty}$ è minore di $v_{rel}$ perché gli asteroidi non vengono attratti dalla gravità terrestre

##### absolute_magnitude
La absolute magnitude (H) è una misura della luminosità di un oggetto celeste; quanto più un oggetto è luminoso (intrinsecamente brillante), tanto minore sarà il suo valore di magnitudine.
La magnitudine assoluta di un asteroide è la magnitudine visuale che un osservatore registrerebbe se l’asteroide fosse posto a 1 Unità Astronomica (au) di distanza, a 1 au dal Sole e con un angolo di fase pari a zero.

##### is_future
boolean: <br>

# Preparazione dei dati

Abbiamo a disposizione due dataset, close_approaches e near_earth; il primo raccoglie i dati relativi appunto ai close_approaches degli asteroidi, cioè quegli asteroidi classificati come near earth asteroids che sono passati o passeranno secondo stime vicino al nostro pianeta, near_earth invece comprende tutti gli asteroidi near earth, quindi anche quelli che non fanno parte di close_approaches.
Il nostro obiettivo è, tramite i dati combinati dei due dataset, sviluppare un modello che classifichi se un asteroide sia potenzailmente pericoloso (pha), e un modello che né calcoli la miss distance. Per fare ciò iniziamo con un merge dei due dataset.

```python
merged["close_approach_date"] = pd.to_datetime(merged["close_approach_date"])
merged["first_obs"] = pd.to_datetime(merged["first_obs"])
merged["last_obs"] = pd.to_datetime(merged["last_obs"])
merged[["is_future", "pha", "diameter_is_estimated"]] = merged[["is_future", "pha", "diameter_is_estimated"]].astype(bool)
size_order = [
    "Tiny (<25m) — Airburst/harmless",
    "Small (25-140m) — Local damage",
    "Medium (140m-1km) — Regional damage",
    "Large (>1 km) — City killer+"
]
merged["size_category"] = pd.Categorical(merged['size_category'], categories=size_order, ordered=True)
merged["class"] = pd.Categorical(merged["class"])
```

```python
merged.info(memory_usage="deep")
```

Verifichiamo che tutti i dati siano validi:

```python
checks = {
    "e": (merged["e"] >= 0) & (merged["e"] < 1),
    "a": merged["a"] > 0,
    "q_ad": merged["q"] <= merged["ad"],
    "moid": merged["moid_au"] >= 0,
    "i": (merged["i"] >= 0) & (merged["i"] <= 180),
    "velocity": merged["velocity_km_s"] > 0,
    "velocity_inf": merged["velocity_infinity_km_s"] > 0,
    "distance_order": ((merged["distance_min_au"] <= merged["distance_au"]) & (merged["distance_au"] <= merged["distance_max_au"])),
    "data_arc": merged["data_arc"] >= 0,
    "obs_order": merged["first_obs"] <= merged["last_obs"],
    "condition_code": merged["condition_code"].between(0, 9)
}

pd.DataFrame(checks).sum()
```

```python
merged.shape
```

Ogni test è risultato positivo, ad eccezione di 9 casi negativi con distance_min_au e distance_max_au.

```python
r = merged['distance_au'] > merged['distance_max_au']
print(merged[r][['distance_min_au', 'distance_au', 'distance_max_au']])
print("righe problematiche:", r.sum())
```

Le 9 eccezioni sono quindi tutte relative a distance_max_au. <br>
Il valore nominale distance_au risulta più basso di distance_max_au per alcune istanze, questo probabilmente è dovuto ad errori da arrotondamenti, infatti l'errore è molto piccolo. <br> Risolviamo portando distance_max_au al valore calcolato in distance_au dove questo è maggiore.

```python
merged['distance_max_au'] = merged[['distance_au', 'distance_max_au']].max(axis=1)
```

### Classificazione dei PHA

```python
corr['pha'].sort_values(ascending=False)
```

Una volta analizzate le correlazioni tra le variabili e il target `pha`, 
utilizziamo un Decision Tree per confermare quali features sono davvero 
discriminanti per la classificazione, e per capire se il modello riesce 
a predire un PHA usando solo parametri orbitali, escludendo le variabili 
che ne definiscono formalmente la classificazione (`absolute_magnitude` e `moid_au`).

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

def eval(model, y, predictions):
    R2 = r2_score(y, predictions)
    mae = mean_absolute_error(y, predictions)
    mse = mean_squared_error(y, predictions)
    rmse = np.sqrt(mse) 
    rmspe = np.sqrt(np.mean((predictions /y - 1) ** 2))
    print("Valutazione del Modello")
    print(f"R²:    {R2:.4f}")
    print(f"MAE:   {mae:.4f}")
    print(f"MSE:   {mse:.4f}")
    print(f"RMSE:  {rmse:.4f}")
    print(f"RMSPE: {rmspe:.4f}")
```

A questo punto per migliorare sensibilmente il modello è necessario inserire variabili specifiche relative al dominio che ci interessa, cioè quello astronomico, non avendo le conoscenze adeguate ci possiamo fermare qua; oppure chiedere ad una LLM delle variabili derivate da quelle già in nostro possesso che consentano al modello di imparare meglio le relazioni orbitali.

```python
def engineer_features(df):
    # Creiamo una copia per non modificare il dataframe originale in-place
    df = df.copy() 
    
    # 1. Prossimità geometrica all'orbita terrestre
    df['diff_a_earth'] = np.abs(df['a'] - 1.0)
    
    # 2. Interazioni non lineari (Velocità in km/s)
    # np.log1p è ideale anche per i km/s poiché gestisce bene i valori piccoli
    log_v = np.log1p(df['velocity_km_s']) 
    
    # Indice di impatto: MOID ponderato per la velocità
    df['potential_impact_index'] = df['moid_au'] / (log_v + 1)
    
    # 3. Parametro di Tisserand "Sicuro"
    # Protezione 1: Evitiamo divisioni per zero sul semiasse maggiore
    safe_a = np.where(df['a'] <= 0, 0.001, df['a'])
    
    # Protezione 2: Gestione orbite iperboliche (e >= 1) per evitare radici negative
    inner_sqrt = safe_a * (1 - df['e']**2)
    inner_sqrt = np.where(inner_sqrt < 0, 0, inner_sqrt) 
    
    # Calcolo del parametro (assumendo l'inclinazione 'i' in gradi)
    df['tisserand'] = (1 / safe_a) + 2 * np.sqrt(inner_sqrt) * np.cos(np.radians(df['i']))
    
    # 4. Pulizia finale: sostituiamo infiniti sfuggiti ai controlli con NaN
    df = df.replace([np.inf, -np.inf], np.nan)
    
    # ATTENZIONE: Nessun fillna(0) applicato. 
    # Affidiamo la gestione dei valori mancanti (NaN) alla logica nativa di XGBoost.
    
    return df
```
