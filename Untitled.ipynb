### Definition of Potentially Hazardous Asteroids

Nasa defines the Potentially Hazardous Asteroids (PHAs) based on parameters that measure the asteroid’s potential to make threatening close approaches to the Earth. Specifically, all asteroids with a minimum orbit intersection distance (MOID) of 0.05 au or less and an absolute magnitude (H) of 22.0 or less, the latter of which roughly corresponds to a size larger than 140 meters, are considered PHAs.

Ulteriore spiegazione delle variabili:

##### designation:
The designation can be temporary (year of discovery, one letter indicating the half-month of the object's discovery within that year and ranges from A (first half of January) to Y (second half of December) and another letter with a numerical suffix indicating the order of discovery within that half-month. The first 25 discoveries of the half-month receive a letter A to Z (excluding the letter I, which is not used to avoid confusion with the digit 1). Because modern techniques typically yield hundreds if not thousands of discoveries per half-month, a subscript number is appended to subsequent discoveries within that half-month to indicate the number of times that the letters A to Z have been cycled through. Thus the suffix 1 indicates discoveries after one completed cycle, namely the 26th through 50th discoveries.  

The permanent designation is a unique numerical sequence. 
<pre>
Example: 
  provisonal designation -> 2005 YQ96 
  permanent  designation -> 613286
</pre>

##### full_name:
The combination of the permanent and provisonal designations
<pre>
Example:
  613286 (2005 YQ96)
</pre>

##### close_approach_date:
The date and time of closest approach in a date format
<pre>
Example: 2015-01-02 21:46
</pre>

##### distance_au, dist_km and dist_lunar:
The miss distance in Astronomical Units (AU) and Kilometers (km), AU is a unit of length defined as exactly equal to 149597870700 m or approximately <br>150 million km, a lunar distance is the distance between Earth and the Moon: 384399 km.
<pre>
example:
    distance_au ->       0.026522
    dist_km     -> 3967664.0
    dist_lunar  ->      10.32
</pre>

##### distance_min_au and distance_max_au
The minimum and maximum possible distance is based on the 3-sigma Earth target-plane error ellipse. <br>
The minimum distance is the closest possible the asteroid could get, given the current errors in our data. <br>
The maximum distance is the farthest possible the asteroid could be during that specific approach. <br>
<pre>
Example:
    asteroid 613286 has distance_au = distance_min_au = distance_max_au.
    That is because we have been able to calculate with a neglectable error its orbit
</pre>



##### velocity_infinity_km_s	

For asteroids it refers to an object’s speed relative to a planet (Earth in this case) or the Sun when it is far enough away to be unaffected by its gravity. <br>
Normally $v_{\infty}$ is lower than $v_{rel}$ because the asteroids doesn't get pulled by earth's gravity

##### absolute_magnitude
absolute magnitude (M) is a measure of the luminosity of a celestial object, the more luminous (intrinsically bright) an object, the lower its magnitude number.
An asteroid’s absolute magnitude is the visual magnitude an observer would record if the asteroid were placed 1 Astronomical Unit (au) away, and 1 au from the Sun and at a zero phase angle.

##### JPL Small-Body SPK-ID
The JPL Small-Body SPK-ID is a unique integer assigned by NASA's Jet Propulsion Laboratory (JPL) to every asteroid, comet, or planet in their database.
It's a different unique number from the permanent designation.

##### full_name
Full name is the combined primary and provisional designation of a celestial body
<pre>
Example:
    433 Eros (A898 PA) => primary designation = 433, proper name = Eros, provisional key = A898 PA
</pre>

##### pdes
Primary Designation is the main identifier used for a body; this can be either the permanent number if the body has been officially numbered, or the provisional designation if it hasn't been numbered yet.
<pre>
    spkid       full_name           pdes
    
    20000433	433 Eros (A898 PA)	433	

    54606826	(2026 FJ4)	        2026 FJ
</pre>

##### name
Proper Name given to bodies which are particularly important, given only if body has been already numbered

<pre>
Example:
    spkid       full_name           pdes        name
    
    20000433	433 Eros (A898 PA)	433	Eros    Eros
    54606826	(2026 FJ4)	        2026 FJ4    NaN

</pre>

##### diameter_km

##### diameter_m

### Checking how many null values are in each dataframe

### Analizing missing data from columns

**Velocity_infinity_km_s** could have null elements due to different factors, the most likely reason as to the absense of these data points is that there simply wasn't enough data to compute, seeing as velocity_km_s has no nulls these missing data points shuoldn't cause too many issues 

**absolute_magnitude** likely has null data points due to a lack of data from the original database

**skpid** can contain missing data points due to the asteroid being recently discovered and not yet in the `near_earth` dataset, a merge logic error, or un update to the database between the two dataset versions

**name** may be missing due to the body not being of importance

**H** [See absolute magnitude in "close_approaches"]

**albedo** is a complex measurement, potentially reserved for more significant bodies

**rot_per** similar to _albedo_ is not a simple mesurement so potentially reserved for more significant bodies

**moid_au/km/lunar_distance** could be due to missing mesurements or a high condition score, this is important because moid is a defining factor when determining PHA classification

**condition_code** could be due to a body being discovered before the implementation of condition codes, data entry errors or other edge cases

**data_arc/years** seeing as only one observation date is missing these missing data points are given by the fact that the first and last observation where the same date

**key_num** asteroid has not yet been numbered, it only holds a provisional name

```python
merged['close_approach_date'] = pd.to_datetime(merged['close_approach_date'])
merged['is_future'] = merged['is_future'].astype(bool)
merged["pha"] = merged["pha"].astype(bool)
merged["diameter_is_estimated"] = merged["diameter_is_estimated"].astype(bool)
size_order = [
    'Tiny (<25m) — Airburst/harmless',
    'Small (25-140m) — Local damage',
    'Medium (140m-1km) — Regional damage',
    'Large (>1 km) — City killer+'
]

merged['size_category'] = pd.Categorical(
    merged['size_category'], 
    categories=size_order, 
    ordered=True
)
merged['class'] = pd.Categorical(merged['class'])
merged['spkid'] = merged['spkid'].astype(np.int64)
```

```python
assert (merged['e'].between(0, 1)).all(), "eccentricity out of range"
assert (merged['i'].between(0, 180)).all(), "inclination out of range"
assert (merged['moid_au'].dropna() >= 0).all(), "moid_au negative"
assert (merged['condition_code'].dropna().between(0, 9)).all(), "condition_code out of range"
```

```python
mask = merged['distance_au'] > merged['distance_max_au']
print(merged[mask][['distance_min_au', 'distance_au', 'distance_max_au']])
print(f"Bad rows: {mask.sum()}")
```

Il valore nominale distance_au risulta più basso di distance_max_au per alcune istanze, questo probabilmente è dovuto ad errori da arrotondamenti, infatti l'errore è molto piccolo. Risolviamo portando distance_max_au al valore calcolato in distance_au dove questo è maggiore.

```python
merged['distance_max_au'] = merged[['distance_au', 'distance_max_au']].max(axis=1)
```
